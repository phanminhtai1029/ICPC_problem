# Problem B:

```python
arr = [1, 3, 5]
k = 3
dict = {}

for x in arr:
    binary_str = bin(x)[2:][::-1]
    for idx, i in enumerate(binary_str):
        if i == "1":
            if idx in dict:
                dict[idx] += 1
            else:
                dict[idx] = 1

max_key = max(dict.keys())

arr1 = []
while any(value != 0 for value in dict.values()):
    string = ""
    for i in range(3, -1, -1):
        if i in dict and dict[i] != 0:
            string += "1"
            dict[i] -= 1
        else:
            string += "0"
    arr1.append(string)

arr2 = []
for i in arr1:
    i = int(i, 2)
    arr2.append(i)

S = 0
for i in range(k):
    S += arr2[i]**2

print(S)
```

# Problem C:

```python
# Input example
n, L = map(int, input().split())
points = [tuple(map(int, input().split())) for _ in range(n)]

# Lấy tất cả tọa độ x và y, phục vụ coordinate compression
xs = sorted(set(x for x, y, w in points) | {x + L for x, y, w in points})
ys = sorted(set(y for x, y, w in points) | {y + L for x, y, w in points})

# Mapping từ giá trị gốc sang index
x_id = {x:i for i, x in enumerate(xs)}
y_id = {y:i for i, y in enumerate(ys)}

# Tạo lưới 2D để tính prefix sum
grid = [[0]* (len(ys)+1) for _ in range(len(xs)+1)]
for x, y, w in points:
    grid[x_id[x]+1][y_id[y]+1] += w

# Tính 2D prefix sum
for i in range(1, len(xs)+1):
    for j in range(1, len(ys)+1):
        grid[i][j] += grid[i-1][j] + grid[i][j-1] - grid[i-1][j-1]

# Hàm tính tổng giá trị trong hình chữ nhật (x1..x2, y1..y2)
def query(x1, y1, x2, y2):
    return grid[x2][y2] - grid[x1][y2] - grid[x2][y1] + grid[x1][y1]

# Tìm giá trị lớn nhất
ans = -10**18
for x in xs:
    for y in ys:
        # xác định index cho (x+L, y+L)
        xi = x_id.get(x+L)
        yi = y_id.get(y+L)
        if xi is not None and yi is not None:
            val = query(x_id[x], y_id[y], xi+1, yi+1)
            ans = max(ans, val)

print(ans)
```

```python
L = 3
points = [(0, 0, 5), (1, 2, -2), (3, 1, 4), (5, 5, 10), (2, 3, 1)]
# Tìm max x
max_x = max(p[0] for p in points)

# Tìm max y
max_y = max(p[1] for p in points)

print("Max x:", max_x)
print("Max y:", max_y)

arr = []
for i in range(max_y+1):
    for j in range(max_x+1):
        S = 0
        for x, y, w in points:
            if j <= x <= j+L and i <= y <= i+L:
                S += w
                arr.append(S)
                
print(max(arr))
```

# Problem D:

```python
m = 10
days = [2, 3]

t = set(range(1, m+1))
print(t)  # Output: (0, 1, 2, 3, 4)

x = set()

for d in days:
    person = set(i for i in range(1, m+1) if i % d == 0)
    x.update(person)

print(len(t - x))
```
